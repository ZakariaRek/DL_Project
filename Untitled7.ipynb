```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from PIL import Image
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ==================== CONFIGURATION ====================
class Config:
    # Chemins - MODIFIER ICI AVEC VOTRE CHEMIN
    DATA_PATH = '/content/orange-diseases-dataset'  # Pour Google Colab
    # DATA_PATH = './orange-diseases-dataset'  # Pour local
    MODEL_SAVE_PATH = 'shufflenet_v2_orange_diseases.pth'

    # Hyperparam√®tres
    BATCH_SIZE = 32
    NUM_EPOCHS = 50
    LEARNING_RATE = 0.001
    IMG_SIZE = 224
    NUM_WORKERS = 4

    # Entra√Ænement
    TRAIN_SPLIT = 0.8
    VAL_SPLIT = 0.1
    TEST_SPLIT = 0.1

    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ==================== DATASET PERSONNALIS√â ====================
class OrangeDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

# ==================== PR√âPARATION DES DONN√âES ====================
def prepare_data(data_path, split='train'):
    """Charge et pr√©pare les donn√©es du dataset
    Args:
        data_path: Chemin vers le dossier principal du dataset
        split: 'train' ou 'test'
    """
    image_paths = []
    labels = []
    class_names = []

    split_path = os.path.join(data_path, split)

    # Parcourir les dossiers de classes dans train/ ou test/
    for class_idx, class_name in enumerate(sorted(os.listdir(split_path))):
        class_path = os.path.join(split_path, class_name)
        if os.path.isdir(class_path):
            if split == 'train':  # On collecte les noms de classes une seule fois
                class_names.append(class_name)
            for img_name in os.listdir(class_path):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_paths.append(os.path.join(class_path, img_name))
                    labels.append(class_idx)

    return np.array(image_paths), np.array(labels), class_names

def get_transforms(train=True):
    """D√©finit les transformations pour l'augmentation de donn√©es"""
    if train:
        return transforms.Compose([
            transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.3),
            transforms.RandomRotation(20),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                               std=[0.229, 0.224, 0.225])
        ])
    else:
        return transforms.Compose([
            transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                               std=[0.229, 0.224, 0.225])
        ])

# ==================== MOD√àLE ====================
def create_model(num_classes):
    """Cr√©e le mod√®le ShuffleNet V2 avec une nouvelle t√™te de classification"""
    # Charger ShuffleNet V2 pr√©-entra√Æn√©
    model = models.shufflenet_v2_x1_0(pretrained=True)

    # Geler les premi√®res couches (optionnel)
    # for param in model.parameters():
    #     param.requires_grad = False

    # Remplacer la derni√®re couche pour le nombre de classes
    num_features = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(num_features, num_classes)
    )

    return model.to(Config.DEVICE)

# ==================== ENTRA√éNEMENT ====================
class Trainer:
    def __init__(self, model, train_loader, val_loader, criterion, optimizer):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.criterion = criterion
        self.optimizer = optimizer
        self.train_losses = []
        self.val_losses = []
        self.train_accs = []
        self.val_accs = []

    def train_epoch(self):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in self.train_loader:
            inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)

            self.optimizer.zero_grad()
            outputs = self.model(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            self.optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        epoch_loss = running_loss / len(self.train_loader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc

    def validate(self):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in self.val_loader:
                inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)
                outputs = self.model(inputs)
                loss = self.criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

        val_loss = running_loss / len(self.val_loader)
        val_acc = 100. * correct / total
        return val_loss, val_acc

    def train(self, num_epochs):
        best_val_acc = 0.0

        for epoch in range(num_epochs):
            train_loss, train_acc = self.train_epoch()
            val_loss, val_acc = self.validate()

            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)

            print(f'Epoch [{epoch+1}/{num_epochs}]')
            print(f'Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%')
            print(f'Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%')
            print('-' * 60)

            # Sauvegarder le meilleur mod√®le
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(self.model.state_dict(), Config.MODEL_SAVE_PATH)
                print(f'‚úì Mod√®le sauvegard√©! (Val Acc: {val_acc:.2f}%)')

# ==================== √âVALUATION ====================
def evaluate_model(model, test_loader, class_names):
    """√âvalue le mod√®le sur le jeu de test"""
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)
            outputs = model(inputs)
            _, predicted = outputs.max(1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Rapport de classification
    print("\n" + "="*60)
    print("RAPPORT DE CLASSIFICATION")
    print("="*60)
    print(classification_report(all_labels, all_preds,
                                target_names=class_names))

    # Matrice de confusion
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Matrice de Confusion')
    plt.ylabel('Vraie Classe')
    plt.xlabel('Classe Pr√©dite')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()

    return all_preds, all_labels

def plot_training_history(trainer):
    """Affiche les courbes d'entra√Ænement"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Loss
    ax1.plot(trainer.train_losses, label='Train Loss')
    ax1.plot(trainer.val_losses, label='Val Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.set_title('√âvolution de la Loss')
    ax1.legend()
    ax1.grid(True)

    # Accuracy
    ax2.plot(trainer.train_accs, label='Train Acc')
    ax2.plot(trainer.val_accs, label='Val Acc')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy (%)')
    ax2.set_title('√âvolution de l\'Accuracy')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

# ==================== PIPELINE PRINCIPAL ====================
def main():
    print("="*60)
    print("CLASSIFICATION DES MALADIES D'ORANGES")
    print("Mod√®le: ShuffleNet V2")
    print("="*60)
    print(f"Device: {Config.DEVICE}\n")

    # 1. Charger les donn√©es
    print("1. Chargement des donn√©es...")

    # Charger train et test s√©par√©ment (le dataset est d√©j√† divis√©)
    X_train, y_train, class_names = prepare_data(Config.DATA_PATH, split='train')
    X_test, y_test, _ = prepare_data(Config.DATA_PATH, split='test')

    print(f"   Train: {len(X_train)} images")
    print(f"   Test: {len(X_test)} images")
    print(f"   Classes: {class_names}")
    print(f"   Nombre de classes: {len(class_names)}\n")

    # 2. Cr√©er un split de validation √† partir du train
    print("2. Cr√©ation du split de validation...")
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.15,  # 15% pour validation
        stratify=y_train, random_state=42
    )
    print(f"   Train (final): {len(X_train)} images")
    print(f"   Val: {len(X_val)} images")
    print(f"   Test: {len(X_test)} images\n")

    # 3. Cr√©er les datasets et dataloaders
    print("3. Cr√©ation des DataLoaders...")
    train_dataset = OrangeDiseaseDataset(X_train, y_train, get_transforms(train=True))
    val_dataset = OrangeDiseaseDataset(X_val, y_val, get_transforms(train=False))
    test_dataset = OrangeDiseaseDataset(X_test, y_test, get_transforms(train=False))
    train_dataset = OrangeDiseaseDataset(X_train, y_train, get_transforms(train=True))
    val_dataset = OrangeDiseaseDataset(X_val, y_val, get_transforms(train=False))
    test_dataset = OrangeDiseaseDataset(X_test, y_test, get_transforms(train=False))

    train_loader = DataLoader(train_dataset, batch_size=Config.BATCH_SIZE,
                             shuffle=True, num_workers=Config.NUM_WORKERS)
    val_loader = DataLoader(val_dataset, batch_size=Config.BATCH_SIZE,
                           shuffle=False, num_workers=Config.NUM_WORKERS)
    test_loader = DataLoader(test_dataset, batch_size=Config.BATCH_SIZE,
                            shuffle=False, num_workers=Config.NUM_WORKERS)

    # 4. Cr√©er le mod√®le
    print("4. Cr√©ation du mod√®le ShuffleNet V2...")
    model = create_model(num_classes=len(class_names))
    print(f"   Nombre de param√®tres: {sum(p.numel() for p in model.parameters()):,}\n")

    # 5. D√©finir loss et optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=Config.LEARNING_RATE)

    # 6. Entra√Æner le mod√®le
    print("5. D√©but de l'entra√Ænement...\n")
    trainer = Trainer(model, train_loader, val_loader, criterion, optimizer)
    trainer.train(Config.NUM_EPOCHS)

    # 7. Charger le meilleur mod√®le et √©valuer
    print("\n6. √âvaluation sur le jeu de test...")
    model.load_state_dict(torch.load(Config.MODEL_SAVE_PATH))
    evaluate_model(model, test_loader, class_names)

    # 8. Afficher l'historique d'entra√Ænement
    print("\n7. G√©n√©ration des graphiques...")
    plot_training_history(trainer)

    print("\n" + "="*60)
    print("ENTRA√éNEMENT TERMIN√â!")
    print("="*60)

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, models
from PIL import Image
import os
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# ==================== CONFIGURATION ====================
class Config:
    # Chemins - MODIFIER ICI AVEC VOTRE CHEMIN
    DATA_PATH = '/content/orange-diseases-dataset'  # Pour Google Colab
    # DATA_PATH = './orange-diseases-dataset'  # Pour local
    MODEL_SAVE_PATH = 'shufflenet_v2_orange_diseases.pth'

    # Hyperparam√®tres
    BATCH_SIZE = 32
    NUM_EPOCHS = 10
    LEARNING_RATE = 0.001
    IMG_SIZE = 224
    NUM_WORKERS = 4

    # Entra√Ænement
    TRAIN_SPLIT = 0.8
    VAL_SPLIT = 0.1
    TEST_SPLIT = 0.1

    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ==================== DATASET PERSONNALIS√â ====================
class OrangeDiseaseDataset(Dataset):
    def __init__(self, image_paths, labels, transform=None):
        self.image_paths = image_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

# ==================== PR√âPARATION DES DONN√âES ====================
def prepare_data(data_path, split='train'):
    """Charge et pr√©pare les donn√©es du dataset
    Args:
        data_path: Chemin vers le dossier principal du dataset
        split: 'train' ou 'test'
    """
    image_paths = []
    labels = []
    class_names = []

    split_path = os.path.join(data_path, split)

    # Parcourir les dossiers de classes dans train/ ou test/
    for class_idx, class_name in enumerate(sorted(os.listdir(split_path))):
        class_path = os.path.join(split_path, class_name)
        if os.path.isdir(class_path):
            if split == 'train':  # On collecte les noms de classes une seule fois
                class_names.append(class_name)
            for img_name in os.listdir(class_path):
                if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_paths.append(os.path.join(class_path, img_name))
                    labels.append(class_idx)

    return np.array(image_paths), np.array(labels), class_names

def get_transforms(train=True):
    """D√©finit les transformations pour l'augmentation de donn√©es"""
    if train:
        return transforms.Compose([
            transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.3),
            transforms.RandomRotation(20),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                               std=[0.229, 0.224, 0.225])
        ])
    else:
        return transforms.Compose([
            transforms.Resize((Config.IMG_SIZE, Config.IMG_SIZE)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                               std=[0.229, 0.224, 0.225])
        ])

# ==================== MOD√àLE ====================
def create_model(num_classes):
    """Cr√©e le mod√®le ShuffleNet V2 avec une nouvelle t√™te de classification"""
    # Charger ShuffleNet V2 pr√©-entra√Æn√©
    model = models.shufflenet_v2_x1_0(pretrained=True)

    # Geler les premi√®res couches (optionnel)
    # for param in model.parameters():
    #     param.requires_grad = False

    # Remplacer la derni√®re couche pour le nombre de classes
    num_features = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(num_features, num_classes)
    )

    return model.to(Config.DEVICE)

# ==================== ENTRA√éNEMENT ====================
class Trainer:
    def __init__(self, model, train_loader, val_loader, criterion, optimizer):
        self.model = model
        self.train_loader = train_loader
        self.val_loader = val_loader
        self.criterion = criterion
        self.optimizer = optimizer
        self.train_losses = []
        self.val_losses = []
        self.train_accs = []
        self.val_accs = []

    def train_epoch(self):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, labels in self.train_loader:
            inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)

            self.optimizer.zero_grad()
            outputs = self.model(inputs)
            loss = self.criterion(outputs, labels)
            loss.backward()
            self.optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        epoch_loss = running_loss / len(self.train_loader)
        epoch_acc = 100. * correct / total
        return epoch_loss, epoch_acc

    def validate(self):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, labels in self.val_loader:
                inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)
                outputs = self.model(inputs)
                loss = self.criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

        val_loss = running_loss / len(self.val_loader)
        val_acc = 100. * correct / total
        return val_loss, val_acc

    def train(self, num_epochs):
        best_val_acc = 0.0

        for epoch in range(num_epochs):
            train_loss, train_acc = self.train_epoch()
            val_loss, val_acc = self.validate()

            self.train_losses.append(train_loss)
            self.val_losses.append(val_loss)
            self.train_accs.append(train_acc)
            self.val_accs.append(val_acc)

            print(f'Epoch [{epoch+1}/{num_epochs}]')
            print(f'Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%')
            print(f'Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%')
            print('-' * 60)

            # Sauvegarder le meilleur mod√®le
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                torch.save(self.model.state_dict(), Config.MODEL_SAVE_PATH)
                print(f'‚úì Mod√®le sauvegard√©! (Val Acc: {val_acc:.2f}%)')

# ==================== √âVALUATION ====================
def evaluate_model(model, test_loader, class_names):
    """√âvalue le mod√®le sur le jeu de test"""
    model.eval()
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in test_loader:
            inputs, labels = inputs.to(Config.DEVICE), labels.to(Config.DEVICE)
            outputs = model(inputs)
            _, predicted = outputs.max(1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Rapport de classification
    print("\n" + "="*60)
    print("RAPPORT DE CLASSIFICATION")
    print("="*60)
    print(classification_report(all_labels, all_preds,
                                target_names=class_names))

    # Matrice de confusion
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Matrice de Confusion')
    plt.ylabel('Vraie Classe')
    plt.xlabel('Classe Pr√©dite')
    plt.tight_layout()
    plt.savefig('confusion_matrix.png')
    plt.show()

    return all_preds, all_labels

def plot_training_history(trainer):
    """Affiche les courbes d'entra√Ænement"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Loss
    ax1.plot(trainer.train_losses, label='Train Loss')
    ax1.plot(trainer.val_losses, label='Val Loss')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.set_title('√âvolution de la Loss')
    ax1.legend()
    ax1.grid(True)

    # Accuracy
    ax2.plot(trainer.train_accs, label='Train Acc')
    ax2.plot(trainer.val_accs, label='Val Acc')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy (%)')
    ax2.set_title('√âvolution de l\'Accuracy')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

# ==================== PIPELINE PRINCIPAL ====================
def main():
    print("="*60)
    print("CLASSIFICATION DES MALADIES D'ORANGES")
    print("Mod√®le: ShuffleNet V2")
    print("="*60)
    print(f"Device: {Config.DEVICE}\n")

    # 1. Charger les donn√©es
    print("1. Chargement des donn√©es...")

    # Charger train et test s√©par√©ment (le dataset est d√©j√† divis√©)
    X_train, y_train, class_names = prepare_data(Config.DATA_PATH, split='train')
    X_test, y_test, _ = prepare_data(Config.DATA_PATH, split='test')

    print(f"   Train: {len(X_train)} images")
    print(f"   Test: {len(X_test)} images")
    print(f"   Classes: {class_names}")
    print(f"   Nombre de classes: {len(class_names)}\n")

    # 2. Cr√©er un split de validation √† partir du train
    print("2. Cr√©ation du split de validation...")
    X_train, X_val, y_train, y_val = train_test_split(
        X_train, y_train, test_size=0.15,  # 15% pour validation
        stratify=y_train, random_state=42
    )
    print(f"   Train (final): {len(X_train)} images")
    print(f"   Val: {len(X_val)} images")
    print(f"   Test: {len(X_test)} images\n")

    # 3. Cr√©er les datasets et dataloaders
    print("3. Cr√©ation des DataLoaders...")
    train_dataset = OrangeDiseaseDataset(X_train, y_train, get_transforms(train=True))
    val_dataset = OrangeDiseaseDataset(X_val, y_val, get_transforms(train=False))
    test_dataset = OrangeDiseaseDataset(X_test, y_test, get_transforms(train=False))
    train_dataset = OrangeDiseaseDataset(X_train, y_train, get_transforms(train=True))
    val_dataset = OrangeDiseaseDataset(X_val, y_val, get_transforms(train=False))
    test_dataset = OrangeDiseaseDataset(X_test, y_test, get_transforms(train=False))

    train_loader = DataLoader(train_dataset, batch_size=Config.BATCH_SIZE,
                             shuffle=True, num_workers=Config.NUM_WORKERS)
    val_loader = DataLoader(val_dataset, batch_size=Config.BATCH_SIZE,
                           shuffle=False, num_workers=Config.NUM_WORKERS)
    test_loader = DataLoader(test_dataset, batch_size=Config.BATCH_SIZE,
                            shuffle=False, num_workers=Config.NUM_WORKERS)

    # 4. Cr√©er le mod√®le
    print("4. Cr√©ation du mod√®le ShuffleNet V2...")
    model = create_model(num_classes=len(class_names))
    print(f"   Nombre de param√®tres: {sum(p.numel() for p in model.parameters()):,}\n")

    # 5. D√©finir loss et optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=Config.LEARNING_RATE)

    # 6. Entra√Æner le mod√®le
    print("5. D√©but de l'entra√Ænement...\n")
    trainer = Trainer(model, train_loader, val_loader, criterion, optimizer)
    trainer.train(Config.NUM_EPOCHS)

    # 7. Charger le meilleur mod√®le et √©valuer
    print("\n6. √âvaluation sur le jeu de test...")
    model.load_state_dict(torch.load(Config.MODEL_SAVE_PATH))
    evaluate_model(model, test_loader, class_names)

    # 8. Afficher l'historique d'entra√Ænement
    print("\n7. G√©n√©ration des graphiques...")
    plot_training_history(trainer)

    print("\n" + "="*60)
    print("ENTRA√éNEMENT TERMIN√â!")
    print("="*60)

if __name__ == "__main__":
    main()
```

```python
import torch
import torch.nn as nn
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files
from IPython.display import display, Image as IPImage, HTML

# ==================== CONFIGURATION ====================
MODEL_PATH = 'shufflenet_v2_orange_diseases.pth'
IMG_SIZE = 224
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
CLASS_NAMES = ['blackspot', 'canker', 'fresh', 'grenning']

# ==================== FONCTIONS ====================
def load_model():
    """Charge le mod√®le entra√Æn√©"""
    model = models.shufflenet_v2_x1_0(pretrained=False)
    num_features = model.fc.in_features
    model.fc = nn.Sequential(
        nn.Dropout(0.5),
        nn.Linear(num_features, len(CLASS_NAMES))
    )
    model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
    model.to(DEVICE)
    model.eval()
    return model

def get_transform():
    """Transformations pour la pr√©diction"""
    return transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                           std=[0.229, 0.224, 0.225])
    ])

def predict_image(image_path, model, transform):
    """Pr√©dit la classe d'une image"""
    image = Image.open(image_path).convert('RGB')
    image_tensor = transform(image).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        outputs = model(image_tensor)
        probabilities = torch.nn.functional.softmax(outputs, dim=1)
        confidence, predicted_idx = torch.max(probabilities, 1)

    predicted_class = CLASS_NAMES[predicted_idx.item()]
    confidence_pct = confidence.item() * 100
    all_probs = probabilities.cpu().numpy()[0] * 100

    return predicted_class, confidence_pct, all_probs

def visualize_prediction(image_path, predicted_class, confidence, all_probs):
    """Affiche l'image avec les pr√©dictions"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

    # Image
    image = Image.open(image_path)
    ax1.imshow(image)
    ax1.axis('off')
    ax1.set_title(f'Pr√©diction: {predicted_class.upper()}\nConfiance: {confidence:.2f}%',
                  fontsize=14, fontweight='bold',
                  color='green' if confidence > 90 else 'orange')

    # Graphique des probabilit√©s
    colors = ['green' if i == CLASS_NAMES.index(predicted_class) else 'steelblue'
              for i in range(len(CLASS_NAMES))]
    bars = ax2.barh(CLASS_NAMES, all_probs, color=colors, alpha=0.7)
    ax2.set_xlabel('Probabilit√© (%)', fontsize=12)
    ax2.set_title('Distribution des probabilit√©s', fontsize=14, fontweight='bold')
    ax2.set_xlim(0, 100)

    for i, (bar, prob) in enumerate(zip(bars, all_probs)):
        ax2.text(prob + 2, i, f'{prob:.1f}%',
                va='center', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()

# ==================== INTERFACE INTERACTIVE ====================
def interactive_predictor():
    """Interface interactive pour la pr√©diction"""
    print("üçä CLASSIFICATEUR DE MALADIES D'ORANGES")
    print("="*60)
    print("üì§ Uploadez une image...")

    uploaded = files.upload()

    if not uploaded:
        print("‚ùå Aucune image upload√©e!")
        return

    image_name = list(uploaded.keys())[0]

    # Sauvegarder
    with open(image_name, 'wb') as f:
        f.write(uploaded[image_name])

    # Afficher l'image upload√©e
    display(HTML("<h3>üì∏ Image upload√©e:</h3>"))
    display(IPImage(filename=image_name, width=300))

    # Charger le mod√®le
    print("\nüì¶ Chargement du mod√®le...")
    model = load_model()
    transform = get_transform()
    print("‚úì Mod√®le charg√©!\n")

    # Pr√©dire
    print(f"üîç Analyse de l'image: {image_name}\n")
    predicted_class, confidence, all_probs = predict_image(image_name, model, transform)

    # Afficher les r√©sultats
    print("="*60)
    print("R√âSULTAT")
    print("="*60)
    print(f"üçä Classe pr√©dite: {predicted_class.upper()}")
    print(f"üìä Confiance: {confidence:.2f}%")
    print("\nüìà Probabilit√©s par classe:")
    for class_name, prob in zip(CLASS_NAMES, all_probs):
        bar = "‚ñà" * int(prob / 2)
        print(f"   {class_name:12s}: {prob:5.2f}% {bar}")
    print("="*60)

    # Message final
    if confidence > 95:
        print("\n‚úÖ Haute confiance - R√©sultat tr√®s fiable!")
    elif confidence > 85:
        print("\n‚ö†Ô∏è Confiance moyenne - R√©sultat probable")
    else:
        print("\n‚ùå Faible confiance - V√©rifier manuellement")

    # Visualiser
    print("\nüìä G√©n√©ration du graphique...\n")
    visualize_prediction(image_name, predicted_class, confidence, all_probs)

    return predicted_class, confidence

# ==================== LANCER ====================
print("‚úÖ Code charg√©! Ex√©cutez maintenant: interactive_predictor()")
```
